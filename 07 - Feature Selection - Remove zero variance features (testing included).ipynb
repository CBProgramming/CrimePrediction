<h3>Remove features with zero variance<h3>

```python
from sklearn.feature_selection import VarianceThreshold
threshold = 0.0;
variance_threshold = VarianceThreshold(threshold = threshold) #.0005
```

```python
#Get columns with zero variance
def get_zero_vairance_columns(df):
    try:
        variance_threshold.fit(df)
        zero_var_columns = [col for col in df
                            if col not in df.columns[variance_threshold.get_support()]]
    except:
        zero_var_columns = df.columns
    print(len(zero_var_columns),"features have",threshold,"variance.")
    return zero_var_columns
```

```python
zero_var_columns = get_zero_vairance_columns(x_train)
```

```python
#Drop all columns with threshold variance
def drop_zero_variance_columns(df,zero_var_columns):
    df = df.drop(zero_var_columns[:-1], axis = 1)
    return df
```

```python
x_train = drop_zero_variance_columns(x_train, zero_var_columns)
```

```python
x_train.to_csv('x_train_zero_variance_removed.csv', index = False)
```

<h3> Testing begins here <h3>

```python
# assert zero variance removal works as intended
def test_zero_variance_check(filename, expected_remaining_columns):
    df = pd.read_csv(filename)
    original_columns = df.columns.tolist()
    expected_zero_var_columns = [x for x in original_columns if x not in expected_remaining_columns]
    zero_var_columns = get_zero_vairance_columns(df)
    assert len(zero_var_columns) == len(expected_zero_var_columns) +1 or (len(zero_var_columns) == 0 and len(expected_zero_var_columns) == 0), "Length of zero variance columns is different to expected"
    for col in zero_var_columns[:-1]:
        assert col in expected_zero_var_columns, "Zero var columns not as expected." + str(col)
    for col in expected_zero_var_columns:
        assert col in zero_var_columns, "Zero var columns not as expected."
    new_df = drop_zero_variance_columns(df,zero_var_columns)
    assert len(df) == len(new_df), "Dataframe lengths vary pre and post drop columns."
    new_df_columns = new_df.columns.tolist()
    assert len(new_df_columns) == len(expected_remaining_columns), "Post column drop datafram length not as expected"
    for col in new_df_columns:
        assert col in expected_remaining_columns, "Remaining columns not as expected."
    for col in expected_remaining_columns:
        assert col in new_df_columns, "Zero var columns not as expected."
    print("All tests completed successfully")
```

```python
# test dataset where all features have zero variance
filename = 'x_train_all_zero_variance_test.csv'
expected_remaining_columns = ['Wednesday']
test_zero_variance_check(filename, expected_remaining_columns)
```

```python
# test dataset where all but one feature has zero variance
filename = 'x_train_all_but_reports_one_day_ago_zero_variance_test.csv'
expected_remaining_columns = ['Reports 1 day ago','Wednesday']
test_zero_variance_check(filename, expected_remaining_columns)
```
