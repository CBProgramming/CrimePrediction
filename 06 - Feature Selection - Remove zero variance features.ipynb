```python
import sklearn
import pandas as pd
```

```python
x_train = pd.read_csv('x_train.csv')
```

```python
x_train.head()
```

<h3>Remove features with zero variance<h3>

```python
from sklearn.feature_selection import VarianceThreshold
threshold = 0.0;
variance_threshold = VarianceThreshold(threshold = threshold) #.0005
variance_threshold.fit(x_train)
```

```python
variance_threshold.get_support()
```

Get columns with zero variance

```python
zero_var_columns = [col for col in x_train
                   if col not in x_train.columns[variance_threshold.get_support()]]
print(len(zero_var_columns),"features have",threshold,"variance.")
```

Drop all columns with threshold variance

```python
x_train.drop(zero_var_columns, axis = 1)
```

```python
x_train.to_csv('x_train_zero_variance_removed.csv', index = False)
```
