```python
import pandas as pd
import datetime
from datetime import date
import numpy as np
```

```python
# Create dummy dataset for use with GUI
```

```python
x = pd.read_csv('x_test.csv')
y = pd.read_csv('y_test.csv')
```

```python
x.head()
```

```python
df = pd.read_csv('SF_neighborhoods.csv')
neighbourhoods = []
num_neighbourhoods = len(df.index)
for i in range(0,num_neighbourhoods):
    neighbourhoods.append(df.iloc[i,2])
```

```python
days = ['Friday','Saturday','Sunday','Thursday','Tuesday','Wednesday']
```

```python
y
```

```python
# merge x and y data as both required for GUI
merged = pd.concat([x,y], axis = 1, join = "inner")
merged
```

```python
# generate recent dates for dataset
num_neighbourhoods = len(neighbourhoods)
num_records = len(merged.index)
num_dates = num_records/num_neighbourhoods
num_dates
```

```python
current_date = date.today()
current_date_array = []
i = 0
while i < num_neighbourhoods:
    current_date_array.append(current_date)
    i+=1
```

```python
dates = pd.DataFrame()
final_dates = pd.DataFrame()
dates['Date'] = current_date_array
final_dates['Date'] = []
i = 1
while i < num_dates:
    final_dates = final_dates.append(dates)
    current_date += datetime.timedelta(days=1)
    dates['Date'] = current_date
    i+=1
final_dates.reset_index(drop=True, inplace=True)
final_dates
```

```python
# add new dates to gui dataset
merged = pd.concat([merged,final_dates], axis = 1, join = "inner")
merged
```

```python
# generate new neighbourhoods for dataset in balanced quantity
neighbourhood_data = pd.DataFrame()
neighbourhood_data['Neighborhood'] = neighbourhoods
final_neighbourhoods = pd.DataFrame()
final_neighbourhoods['Neighborhood'] = neighbourhoods
i = 1
while i < num_dates:
    final_neighbourhoods = final_neighbourhoods.append(neighbourhood_data)
    i+=1
final_neighbourhoods.reset_index(drop=True, inplace=True)
final_neighbourhoods
```

```python
# add neighbourhoods to GUI dataset
merged = pd.concat([merged,final_neighbourhoods], axis = 1, join = "inner")
merged
```

```python
merged.to_csv('data.csv', index = False)
```
